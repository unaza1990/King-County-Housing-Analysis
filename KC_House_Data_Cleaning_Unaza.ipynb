# Importing libraries 

```python
import pandas as pd
```

# Importing Data

```python
# Load data

df = pd.read_csv("/Users/muhammaddildar/Downloads/kc_house_data.csv")

df.head()
```

```python
# Drop ID column
df = df.drop(columns=['id'])
```

```python
# Convert 'date' column to datetime format for time-based analysis
df['date'] = pd.to_datetime(df['date'])
```

```python
# Convert 'zipcode' to string type to treat it as a categorical variable
df['zipcode'] = df['zipcode'].astype(str)
```

```python
# Create a new column 'was_renovated' to flag homes that have been renovated (1 = renovated, 0 = not renovated)
df['was_renovated'] = df['yr_renovated'].apply(lambda x: 1 if x > 0 else 0)
```

```python
print("Duplicate rows:", df.duplicated().sum())
```

```python
print("Missing values:\n", df.isnull().sum())
```
